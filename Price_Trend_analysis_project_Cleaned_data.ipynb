```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('/content/Nigeria_Agricultural_Price_Data_Mini.xlsx')
```

```python
pd.set_option('display.max_columns',None)
```

```python
df.head(10)
```

```python
# Standardizing the Price column
```

```python
df['Price'] = (df['Price']
    .astype(str)
    .str.replace('₦', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace(r'\.\d+', '', regex=True)
    .str.strip())
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Price']=df['Price'].fillna(0)
```

```python
df.to_excel('Nigeria_Agricultural_Price_Data_Mini_cleaned.xlsx',index=False)
```
